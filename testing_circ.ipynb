```python
import lightworks as lw
```

```python
circ = lw.Circuit(5)
#circ.add(lw.Unitary(lw.random_unitary(3)), 1)
circ.add_ps(0, 1)
circ.add_ps(0, 1)
circ.add_ps(0, 1)
circ.add_ps(0, 1)
circ.add_herald(1, 1, 3)

circ.add_barrier()

circ2 = lw.Circuit(4)
#circ2.add_herald(1, 0, 0)
circ2.add_herald(1, 2, 1)
circ2.add_herald(1, 3, 2)
circ2.add_bs(0)
circ2.add_bs(1)
circ2.add_bs(2)

circ.add(circ2, 0)

circ.add(lw.Unitary(lw.random_unitary(3)), 0)

circ3 = lw.Circuit(3)
circ3.add_bs(0)
circ3.add_bs(1)

circ.add(circ3, 0, group=True)

circ.add_mode_swaps({2:3,3:4,4:2})
circ.add(circ2, 1)

circ.display(display_type = "svg")
```

```python
circ = lw.Circuit(6)

sub_circ = lw.Circuit(5)

sub_circ.add_ps(0, 0)
sub_circ.add_ps(1, 1)
sub_circ.add_ps(2, 2)
sub_circ.add_ps(3, 3)
sub_circ.add_ps(4, 4)
sub_circ.add_herald(1, 0, 3)
sub_circ.add_herald(2, 1, 1)

circ.add(sub_circ, 2)

circ.add_barrier()

circ.add(sub_circ, 1)

circ.add_barrier()

circ.add(sub_circ, 2)

circ.display()

circ.unpack_groups()

circ.display()
```
